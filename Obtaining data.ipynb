# Data Cleaning

```python
# creating a dataset
url_link1 = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
import pandas as pd
company_data = pd.read_html(url_link1)
```

```python
type(company_data)
```

```python
company_data = company_data[0] # dataframe
```

```python
company_data.head()
```

```python
company_data.columns
```

```python
company_data.columns = company_data.columns.get_level_values(0)
```

```python
company_data.columns
```

```python
company_data.head()
```

```python
# Dropping of Ref columns
company_data.drop(['Ref.'], axis = 1,)
```

```python
company_data.head()
```

```python
company_data.drop(['Ref.'], axis = 1, inplace = True)
```

```python
company_data.head()
```

```python
# Taking close look at the Revenue colum
company_data['Revenue'].head(3)
```

```python
# Remove $ sign from revenue column
company_data['Revenue'] = company_data['Revenue'].str.replace('$','')
```

```python
company_data.head(3)
```

```python
# concised data summary
company_data.info()
```

```python
type(company_data.info)
```

```python
# Remove $ sign from Profit column
company_data['Profit'] = company_data['Profit'].str.replace('$','')
```

```python
company_data.head(4)
```

```python
company_data.columns
```

```python
company_data.rename(columns={
    "Revenue": "Revenue($)", 
    "Profit": "Profit($)"
}, inplace=True)
```

```python
company_data.head(5)
```

```python
company_data.columns
```

```python
company_data['Profit($)'].unique()
```

```python
#company_data[[
  #  "Revenue($)","Profit($)"]] = company_data[[
 #   "Revenue($)", "Profit($)"]]
#.astype(float)
```

```python
company_data['Revenue($)'].unique()
```

```python
company_data['Profit($)'].unique()
```

```python
company_data['Profit($)'] = company_data['Profit($)'].str.replace('–','')
```

```python
company_data['Profit($)'].unique()
```

```python
type(company_data['Profit($)'].unique())
```

```python
company_data['Profit($)'] = company_data['Profit($)'].str.replace('...','unknown')
```

```python
company_data['Profit($)'].unique()
```

```python
company_data['Profit($)'] = company_data['Profit($)'].str.replace('('')','')
```

```python
company_data['Profit($)'].unique()
```

```python
company_data['Revenue($)'] = company_data['Revenue($)'].str.replace(',','')
```

```python
company_data['Revenue($)'].unique()
```

```python
company_data['Profit($)'].unique()
```

```python
company_data['Profit($)'] = company_data['Profit($)'].str.replace("''","")
```

```python
company_data['Profit($)'].unique()
```

```python
company_data[[
    "Revenue($)",
    "Profit($)"]] = company_data[["Revenue($)", 
                                  "Profit($)"]].astype(float)
```
